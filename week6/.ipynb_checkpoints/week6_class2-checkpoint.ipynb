```python
a = "wednesday"
a[::-1]
```

```python
i = 0
#generates all 3 digit numbers
for a in range(99, 1000):
    #generates second factor, also 3 digit
    for b in range(99,1000):
        #multiply the numbers and make product a string, because to do [::1] on c, it needs to be a string
        c = str(a*b)
        #check that c backwards is the same as c forwards
        if(c[::-1] == c):
            #print (c);
            i += 1
            print(c, "Palindrome #", i);
```
